```python
import matplotlib
import matplotlib.pyplot as plt
import cv2
import numpy as py
```

```python
from skimage.filters import sobel
from skimage import data
from skimage import filters
from skimage import color
from skimage.filters import gaussian
from skimage import exposure
from skimage import transform
from skimage import morphology
```

```python
image = data.coins()

plt.imshow(image)
```

```python
sobel_image = filters.sobel(image)
plt.imshow(sobel_image)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,10), sharex=True,
sharey=True)
ax1.imshow(image, cmap=plt.cm.gray)
ax1.set_title('image')
ax1.axis('off')
ax2.imshow(sobel_image, cmap=plt.cm.gray)
ax2.set_title('sobel_image')
ax2.axis('off')
```

```python
img = data.cat()
plt.imshow(img)
```

```python
gaussian_image = filters.gaussian(img)
plt.imshow(gaussian_image)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,10), sharex=True,
sharey=True)
ax1.imshow(img, cmap=plt.cm.gray)
ax1.set_title('img')
ax1.axis('off')
ax2.imshow(gaussian_image, cmap=plt.cm.gray)
ax2.set_title('gaussian_image')
ax2.axis('off')
```

```python
img4 = data.horse()
plt.imshow(img4)
```

```python
rectangle = morphology.rectangle(4,2)
footprint=rectangle
eroded_image = morphology.binary_erosion(img4, footprint = footprint)

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 10), sharex=True,
sharey=True)
ax1.imshow(img4, cmap=plt.cm.gray)
ax1.set_title('img4')
ax1.axis('off')
ax2.imshow(eroded_image, cmap=plt.cm.gray)
ax2.set_title('eroded_image')
ax2.axis('off')
```
